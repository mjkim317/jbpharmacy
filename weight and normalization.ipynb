```python
import pandas as pd
import geopandas
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn import preprocessing
```

### 부안

```python
buan = pd.read_csv("C:/Users/user/Desktop/mjpharm/data/buan_normal.csv", encoding='cp949')
buan.head(3)

si = buan['시']
eup = buan['읍면동']
buan = buan.drop(['시','읍면동'], axis=1)
print(buan.dtypes)

scaler = preprocessing.MinMaxScaler()
names = buan.columns
d = scaler.fit_transform(buan)
buan_norm = pd.DataFrame(d, columns=names)
buan_norm

buan_norm['weight'] = buan_norm.sum(axis=1)

sample = pd.concat([si, eup], axis=1)
buan_weight = pd.concat([sample, buan_norm], axis=1)
buan_weight

buan_weight.to_csv("C:/Users/user/Desktop/mjpharm/data/weight/buan_weight.csv")
```
